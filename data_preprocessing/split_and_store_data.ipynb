```python
import os 
import shutil
import random
```

```python
data_dir = 'path of paddy dataset after resizing' 
output_dir = 'path where you want the train/val folders'
split_ratio = 0.8  # 80% for training

# Get all class names (folder names)
classes = [d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))]

for cls in classes:
    # Path to the current class folder
    cls_path = os.path.join(data_dir, cls)
    
    # Get all images in this class folder
    images = [f for f in os.listdir(cls_path) if os.path.isfile(os.path.join(cls_path, f))]
    
    # Shuffle the images to ensure randomness
    random.shuffle(images)
    
    # Calculate split index
    train_count = int(len(images) * split_ratio)
    
    train_images = images[:train_count]
    val_images = images[train_count:]
    
    # 2. Define and create destination folders
    for folder_type, images_to_copy in [('train', train_images), ('val', val_images)]:
        target_dir = os.path.join(output_dir, folder_type, cls)
        os.makedirs(target_dir, exist_ok=True)
        
        # 3. Copy files to disk
        for img in images_to_copy:
            src_path = os.path.join(cls_path, img)
            dst_path = os.path.join(target_dir, img)
            shutil.copy(src_path, dst_path)

print(f"Stratified split complete! Data saved in {output_dir}")
```
